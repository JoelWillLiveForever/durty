```python
#  scikit-learn   train_test_split,          .  
#    75%     .  25%     
# .   ,        ,  ,  
#  ,  25% ,   
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
X, y = mglearn.datasets.make_wave(n_samples=666)
#    wave     
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
#         3
reg = KNeighborsRegressor(n_neighbors=3)
#         
reg.fit(X_train, y_train)
```

```python
print("   :\n{}".format(reg.predict(X_test)))
```

```python
print("R^2   : {:.2f}".format(reg.score(X_test, y_test)))
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
#  1000  ,    -3  3
line = np.linspace(-3, 3, 1000).reshape(-1, 1)
for n_neighbors, ax in zip([1, 3, 9], axes):
#  ,  1, 3,  9 
  reg = KNeighborsRegressor(n_neighbors=n_neighbors)
  reg.fit(X_train, y_train)
  ax.plot(line, reg.predict(line))
  ax.plot(X_train, y_train, '^', c=mglearn.cm2(0), markersize=8)
  ax.plot(X_test, y_test, 'v', c=mglearn.cm2(1), markersize=8)
  ax.set_title(
    "{} neighbor(s)\n train score: {:.2f} test score: {:.2f}".format(
  n_neighbors, reg.score(X_train, y_train),
  reg.score(X_test, y_test)))
  ax.set_xlabel("")
  ax.set_ylabel(" ")

axes[0].legend([" ", " /",
" /"], loc="best");
```

```python
from sklearn.linear_model import LinearRegression
X, y = mglearn.datasets.make_wave(n_samples=666)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
lr = LinearRegression().fit(X_train, y_train)
```

```python
print("lr.coef_: {}".format(lr.coef_))
print("lr.intercept_: {}".format(lr.intercept_))
```

```python
print("R-   : {:.2f}".format(lr.score(X_train, y_train)))
print("R-   : {:.2f}".format(lr.score(X_test, y_test)))
```

```python
from mglearn.datasets import load_extended_boston

X, y = load_extended_boston()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
lr = LinearRegression().fit(X_train, y_train)
```

```python
print("R-   : {:.2f}".format(lr.score(X_train, y_train)))
print("R-   : {:.2f}".format(lr.score(X_test, y_test)))
```

## 

```python
from sklearn.linear_model import Ridge
ridge = Ridge().fit(X_train, y_train)
print("R-   : {:.2f}".format(ridge.score(X_train, y_train)))
print("R-   : {:.2f}".format(ridge.score(X_test, y_test)))
```

```python
ridge10 = Ridge(alpha=10).fit(X_train, y_train)
print("R-   : {:.2f}".format(ridge10.score(X_train, y_train)))
print("R-   : {:.2f}".format(ridge10.score(X_test, y_test)))
```

```python
ridge01 = Ridge(alpha=0.1).fit(X_train, y_train)
print("R-   : {:.2f}".format(ridge01.score(X_train, y_train)))
print("R-   : {:.2f}".format(ridge01.score(X_test, y_test)))
```

```python
plt.plot(ridge.coef_, 's', label="  alpha=1")
plt.plot(ridge10.coef_, '^', label="  alpha=10")
plt.plot(ridge01.coef_, 'v', label="  alpha=0.1")
plt.plot(lr.coef_, 'o', label=" ")
plt.xlabel(" ")
plt.ylabel(" ")
plt.hlines(0, 0, len(lr.coef_))
plt.ylim(-25, 25)
plt.legend();
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso().fit(X_train, y_train)
print("R-   : {:.2f}".format(lasso.score(X_train, y_train)))
print("R-   : {:.2f}".format(lasso.score(X_test, y_test)))
print("  : {}".format(np.sum(lasso.coef_ != 0)))
```

```python
#    "max_iter",
#    ,    max_iter.
lasso001 = Lasso(alpha=0.01, max_iter=100000).fit(X_train, y_train)
print("R-   : {:.2f}".format(lasso001.score(X_train, y_train)))
print("R-   : {:.2f}".format(lasso001.score(X_test, y_test)))
print("  : {}".format(np.sum(lasso001.coef_ != 0)))

```

```python
lasso00001 = Lasso(alpha=0.0001, max_iter=100000).fit(X_train, y_train)
print("R-   : {:.2f}".format(lasso00001.score(X_train, y_train)))
print("R-   : {:.2f}".format(lasso00001.score(X_test, y_test)))
print("  : {}".format(np.sum(lasso00001.coef_ != 0)))
```

```python
plt.plot(lasso.coef_, 's', label=" alpha=1")
plt.plot(lasso001.coef_, '^', label=" alpha=0.01")
plt.plot(lasso00001.coef_, 'v', label=" alpha=0.0001")
plt.plot(ridge01.coef_, 'o', label="  alpha=0.1")
plt.legend(ncol=2, loc=(0, 1.05))
plt.ylim(-25, 25)
plt.xlabel(" ")
plt.ylabel(" ");
```
